```python
import cv2
import mahotas as mt
import numpy as np
import os
from pathlib import Path
from matplotlib import pyplot as plt
import glob
import matplotlib.image as mpimg
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier 
import watershedalgo
```

```python
p = Path("./dataset/")
```

```python
dirs = p.glob("*jpg")   # "*" is a regular expression which will give everything of extension .jpg in the path p
```

```python
print(dirs) #dirs is a generator object
```

```python
labels = []
for d in dirs:  # here we will get all the files names with .jpg extension
    label = str(d).split("\\")[-1]
    print(label)
```

```python
def extract_features(image): 
    # calculate haralick texture features for 4 types of adjacency
    textures = mt.features.haralick(image) 
    print(textures)
    # take the mean of it and return it
    ht_mean = textures.mean(axis=0)
        
    return ht_mean 
    #return kp 
```

```python
image_data = []
label = []
```

```python
i = 1

for file in glob.glob(os.path.join(p, '*.jpg*') ):
    print("current file is: " + file)
    image = cv2.imread(file)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    img = mpimg.imread(file)
    #watershedalgo.lana(image) 
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY))
    plt.axis("off")
    plt.show()
    features = extract_features(gray)
    image_data.append(features)
    labels.append(file)
```

```python
print ("Training features: {}".format(np.array(image_data).shape))
```
