```python
import cv2
import numpy as np
import os
from myshow import myshow
import glob
import mahotas as mt
import SimpleITK
from watershedalgo import lana
import matplotlib.pyplot as plt
import watershedalgo
import superpixelsegmentation
```

```python
train_path = "C:/Users/saree/Desktop/AI PROJECT/BRATS2015/training/HGG"
train_names = os.listdir(train_path) 
print(train_names)
```

```python
train_features = []
train_labels = []
```

```python
def extract_features(image):
    textures = mt.features.haralick(image)
    ht_mean = textures.mean(axis=0) 
    return ht_mean 
```

```python
i = 1
idxSlice = 50
count = 1

for train_name in train_names:  #PATIENT ITERATE
     cur_path = train_path + "/" + train_name
     cur_label = 1
     i = 1
     count+=1
     c = 1
     for fileName in glob.glob(cur_path + "/*.mha"): 
        print("Processing Image - {} in {}".format(i, fileName))
        
        imgT1Original = SimpleITK.ReadImage(fileName)
        
        image = SimpleITK.GetArrayFromImage(SimpleITK.Tile(imgT1Original[:, :, idxSlice],(2, 1, 0)))
        
        if count%5==0:
            myshow(imgT1Original)
            lana(image)
        
        #seg = cv2.imread("seg.jpg")
        image = np.uint8(image) 
        image = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
        
        
        features = extract_features(image) 
        train_features.append(features)
        train_labels.append(cur_label)
        i += 1 
        c+=1
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC

#clf_svm = LinearSVC(random_state = 9,max_iter=100000) 
clf = RandomForestClassifier(max_depth=2, random_state = 0)
# fit the training data and labels

#clf_svm.fit(train_features, train_labels) 
clf.fit(train_features, train_labels)
```
